```python
# importing necessary modules
import pandas as pd
import os
from google.colab import drive

# mount google drive into google collaboratory
drive.mount('/content/gdrive')

# assign the address of the dataset directory
data_dir = '/content/gdrive/MyDrive/GitHub/Python for Data Engineering'

# change the working directory
os.chdir(data_dir)
```

```python
# read the 'sales.csv' file
sales = pd.read_csv('sales.csv')
```

```python
# initial characterization of the entire dataframe
sales.describe(include = 'all')
```

```python
# show the first 5 rows of the dataframe
sales.head()
```

```python
# remove duplicate rows in the dataframe
sales.drop_duplicates()
```

```python
# this block of code shows the the 'Price' column cannot be expressed in thousands 
sales.Price/1000
```

```python
# from the previous cell, we know that there are values in the 'Price' column that are not of the type int, but are rather strings
# to solve this, we first try to convert all values in the 'Price' column into float data type
sales.Price.apply(float)
```

```python
# However, the previous cell showed a value error stating that the code 'could not convert string to float: '13,000''
# checking the error, one possible solution is to remove all commas in the 'Price' column

# we use this code to remove commas in the 'Price' column and convert the values in this column into float64; float64 is a numpy data type
sales.Price = sales.Price.str.replace(',','').astype('float64')
```

```python
# the previous cell has now worked and we have converted the 'Price' column into float data type; we can express price in thousands
# NOTE TO THE CHECKER: It was not stated in the instructions if this expression was to be saved in the 'sales' dataframe; the instruction was only to 'express price in thousands'
sales.Price/1000
```

```python
# filter the sales dataframe with those customers from the United States who ordered Product3 with payment type done via Visa or Amex; 
# data columns for 'Name', 'City', 'State', and 'Country' were selected as well
empty = sales.loc[((sales.Product == 'Product3') & 
                   (sales.Country == 'United States') & 
                  ((sales.Payment_Type == 'Visa') | 
                   (sales.Payment_Type == 'Amex')))] \
                   .filter(['Name', 'City', 'State', 'Country'])
```

```python
# show the contents of the 'empty' dataframe
empty
```

```python
# save dataframe into a .csv file named 'empty.csv'
empty.to_csv('empty.csv')
```

```python
#Select name, price, city, state, and country data of product 2 bought by customer from New York and paid through Mastercard. Store the selected data in empty1.csv file
empty1 = sales.loc[(sales.Product == 'Product2') & 
                  (sales.City == 'New York') & 
                  (sales.Payment_Type == 'Visa')] \
                  .filter(['Name', 'Price', 'City', 'State', 'Country'])
```

```python
# show the contents of the 'empty1' dataframe
empty1
```

Apprently, there are no customers from New York who ordered Product 2 using their Mastercard

```python
empty1.to_csv('empty1.csv')
```
